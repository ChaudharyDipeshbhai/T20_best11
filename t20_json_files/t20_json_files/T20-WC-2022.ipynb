```python
import pandas as pd
import json
```

```python
with open('t20_wc_match_results.json') as f:
    data = json.load(f)
match = pd.DataFrame(data[0]['matchSummary'])
match
```

```python
match.shape
```

```python
match.rename({'scorecard':'match_id'},axis = 1,inplace = True)
```

```python
match.head()
```

```python
match_id_dict = {}
for index,row in match.iterrows():
    key1 = row["team1"] + ' Vs ' + row["team2"]
    key2 = row["team2"] + 'Vs' + row["team1"]

    match_id_dict[key1] = row["match_id"]
    match_id_dict[key2] = row["match_id"]

match_id_dict
```

**Batting Summary**

```python
with open('t20_wc_batting_summary.json') as f:
    data = json.load(f)

    record = []
    for rec in data:
        record.extend(rec['battingSummary'])
batting = pd.DataFrame(record)
batting.head()
```

```python
batting.head(11)
```

```python
batting["out/not_out"] = batting.dismissal.apply(lambda x:"out" if len(x)>0 else "not_out")
batting.head(11)
```

```python
batting.drop(columns = ["dismissal"],inplace = True)
batting.head()
```

```python
batting['batsmanName'] = batting['batsmanName'].apply(lambda x: x.replace('â€',''))
batting['batsmanName'] = batting['batsmanName'].apply(lambda x: x.replace('\xa0',''))
batting.head(11)
```

```python
batting["match_id"] = batting["match"].map(match_id_dict)
batting.head()
```

```python
batting.to_csv('t20_batting_summary.csv',index = False)
```

```python
match.to_csv('t20_match_result.csv',index = False)
```

**Process Bowling Summary**

```python
with open('t20_wc_bowling_summary.json') as f:
    data = json.load(f)
    all_record = []
    for rec in data:
        all_record.extend(rec['bowlingSummary'])
all_record[:2]
```

```python
bowling = pd.DataFrame(all_record)
print(bowling.shape)
bowling.head()
```

```python
bowling['match_id'] = bowling['match'].map(match_id_dict)
bowling.head()
```

```python
bowling.to_csv('t20_bowling_Summary.csv', index = False)
```
